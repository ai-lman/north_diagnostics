# Caching Probe Data

This example tests the difference in time it takes between loading the probes individually and with caching.

```python
from north_diagnostics import Probe
import time
```

```python
path = "/mnt/c/Users/alec/Data/NORTH/ProbeSim_2025/Hydrogen/DDAQ/"
shot = 9774
probe_numbers = [i + 1 for i in range(50)]
```

```python
# ---- Without Caching ----
print("Testing without caching...")
start_time_no_cache = time.time()
probes_no_cache = [Probe(path, shot, number) for number in probe_numbers]

# Triggering lazy loading for fair timing
for probe in probes_no_cache:
    try:
        _ = probe.current
    except:
        continue

end_time_no_cache = time.time()

# ---- With Caching ----
print("\nTesting with caching...")
start_time_cache = time.time()
probes_cache = [Probe(path, shot, number, caching=True) for number in probe_numbers]

# Triggering lazy loading for fair timing
for probe in probes_cache:
    try:
        _ = probe.current
    except:
        continue

end_time_cache = time.time()

print(f"Time without caching: {end_time_no_cache - start_time_no_cache:.4f} seconds")
print(f"Time with caching: {end_time_cache - start_time_cache:.4f} seconds")
```
