# Probe Positions

This example shows the probe positions for the active and inactive (red) probes in the tokamak cross section. 

```python
from north_diagnostics import Probe
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = "/mnt/c/Users/alec/Data/NORTH/ProbeSim_2025/Hydrogen/DDAQ/"
shot = 9774
```

```python
fig, ax = plt.subplots()

theta = np.linspace(0, 2 * np.pi, 500)
circle_x = 250 + 130 * np.cos(theta)
circle_z = 130 * np.sin(theta)
ax.plot(circle_x, circle_z, 'k')
ax.set_aspect('equal', adjustable='box')

def clr(active):
   return 'w' if active else 'C3'

for i in range(Probe.TOTAL_PROBES):
    probe = Probe(path = path, shot = shot, number = i + 1, caching=True)
    ax.scatter(probe.position['r'], probe.position['z'], marker='o', edgecolors='k', facecolors=clr(probe.active))
    ax.text(probe.position['r'], probe.position['z'] - 10, str(probe.number), color='black', fontsize=8, ha='center', va='center')

ax.set_xlabel('r [mm]')
ax.set_ylabel('z [mm]')

ax.set_title('Probe Positions in NORTH')
```
